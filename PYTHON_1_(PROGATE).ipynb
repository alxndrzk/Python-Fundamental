**BOOLEANS**

(True) adalah nilai tipe data boolean.
Tipe data boolean hanya mempunyai dua nilai, (True) dan (False). True akan muncul jika kondisi dipenuhi, dan False jika tidak.
Perlu diingat bahwa Anda harus menulis huruf pertama (True) dan (False) huru awal dalam huruf kapital

OPERATOR PEMBANDINGAN

x < y untuk **true** jika x lebih kecil dari y

x <= y untuk **true** jika x lebih kecil dari atau sama dengan y

x > y untuk **true** jika x lebih besar dari y

x >= y untuk **true** jika x lebih besar dari atau sama dengan y

x == y untuk **true** jika x sama dengan y

x != y untuk **true** jika tidak setara

```python
print(3==3) #true
print(4==5) #false
print(4+2 >= 6) #true
print(4-2 >= 6) #false

#ex          
# Jika x lebih besar dari 30, cetak 'x lebih besar dari 30'            
x = 10
if x >= 10 :
  print('x lebih besar dari 30')          
                                     
money = 5            
apple_price = 2            
 # Jika money sama dengan atau lebih besar dari apple_price, cetak 'Anda dapat membeli apel'            
if money >= apple_price :            
 print('Anda dapat membeli apel')
```

**MENGGABUNGKAN CONDISIONAL EXPRESSION**



```python
#menggunakan operator and
 #operator logika and
  #true and true = true
  #true and false = false
  #false and true = false
  #false and false = false
hour = 7
if hour>6 and hour<8:
 print("pagi!")

#ATAU
hour= 8
if 6<hour<10:
  print('mandi!')

#menggunakan operator or
 #operator logika or akan menghasilkan true jika salah satu kondisi menampilkan true. berbeda dengan and, di operator or hanya membutuhkan satu kondidsi true
hour = 12
if hour ==11 or hour==12:
  print("tidur!")

#menggunakan operator not
#Dengan menggunakan not, Anda dapat menegaskan suatu kondisi.
#Ini artinya False akan muncul jika conditional expression benar dan True akan muncul jika salah
hour = 12
if not hour>=13:
  print('bukan siang')


#CONTOH CODE
x = 20
# Jika x berkisar antara 10 dan 30 (inklusif), cetak 'x berkisar antara 10 dan 30'
if x>=10 and x<=30:
    print('x berkisar antara 10 dan 30')


y = 60
# Jika y lebih kecil dari 10 atau lebih besar dari 30, cetak 'y lebih kecil dari 10 atau lebih besar dari 30'
if y<10 or y>30:
    print('y lebih kecil dari 10 atau lebih besar dari 30')


z = 55
# Jika z tidak sama dengan 77, print 'z bukan 77'
if not z==77:
    print('z bukan 77')


```

**INPUT**

input() memungkinkan Anda dapat mengisi angka pada console ketika menjalankan code, dan menerima nilai yang dimasukkan.
Dengan menulis variable = input('string yang ingin Anda tampilkan di console'), nilai akan ditentukan ke variable.

```python
input_count = input('mau berapa apel?') # nilai yang diterima input selalu dalam bentuk string
print('anda akan membeli ' + input_count + ' apel')
```

```python
apple_price = 2
input_count = input('mau berapa apel?')
print('anda akan membeli ' + input_count +' apel')
count= int(input_count) #diubah ke integer agar bisa dikalikan di total price
total_price = apple_price * count
print('total belanja anda adalah ' + str(total_price) + ' dolar')
```

**MEMBUAT PROGRAM SEDERHANA MENGHITUNG JUMLAH BELANJA**

```python
apple_price = 2
money = 10
input_count = input('mau berapa apel?')
print('anda akan membeli ' + input_count + ' apel')
count = int(input_count)
total_price = apple_price * count
print('total belaja anda adalah ' + str(total_price) + ' dolar')
if money > total_price:
  print('anda telah membeli ' + input_count + ' apel')
  print('uang anda tinggal ' + str(money - total_price) + " dolar")
elif money == total_price :
  print('anda telah membeli ' + input_count + ' apel')
  print('uang anda telah habis')
else :
   print('anda tidak bisa membeli ' + input_count + ' apel')
   print('karena uang anda tidak cukup')
```
